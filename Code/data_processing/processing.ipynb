# BraTS Preprocessing

Pipeline for the BraTS 2023 T1N volumes

Steps: percentile clip -> crop foreground -> pad to cubic -> resize -> normalize to [-1, 1]

Preprocessing method follows Khader et al. 2023


```python
import sys
from pathlib import Path

sys.path.insert(0, str(Path.cwd()))

from src.nifti import load_nifti
from src.preprocessing import process_brats_dataset
from src.visualize import plot_preprocessing_steps, plot_orthogonal_slices, print_volume_stats


PRESETS = {
    16: (16, 16, 16),
    32: (32, 32, 32),
    48: (48, 48, 48),
    64: (64, 64, 64),
}

PRESET = 32
TARGET_SIZE = PRESETS[PRESET]

INPUT_DIR  = r"C:\Users\Sven\Desktop\Data\brain"
OUTPUT_DIR = rf"C:\Users\Sven\Desktop\Data\Processed\brats\{PRESET}"

print(f"preset {PRESET} -> target size {TARGET_SIZE}")
print(f"input:  {INPUT_DIR}")
print(f"output: {OUTPUT_DIR}")
```

```python
sample_path = next(Path(INPUT_DIR).glob("**/*-t1n.nii.gz"))
volume, meta = load_nifti(str(sample_path))

print(f"sample: {sample_path.name}")
print(f"voxel spacing: {meta['spacing']}")
print_volume_stats(volume, "original")
```

```python
# Shows the full preprocessing pipeline on one example volume
plot_preprocessing_steps(volume, TARGET_SIZE)
```

```python
# Run preprocessing on all 1251 BraTS T1N volumes
written = process_brats_dataset(INPUT_DIR, OUTPUT_DIR, TARGET_SIZE)
```

```python
out_path = written[0]
processed, _ = load_nifti(out_path)
print_volume_stats(processed, "processed")
plot_orthogonal_slices(processed, title=Path(out_path).stem)
```
